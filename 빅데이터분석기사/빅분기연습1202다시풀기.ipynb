2-2.캘리포니아하우징, 주어진 데이터의 첫번쨰 행부터 순서대로 80%까지의 데이터를 훈련데이터로 추출 후, 전체 방 개수 컬럼을 의미하는 'total bedrooms'의 변수의 결측치를 'total_bedrooms'변수의 중앙값으로 대체한 데이터 세트를 구성한다. 결측치 대체 전의 'total_bedrooms' 변수 표준편차 값과 결측치 대체 후의 표준편차 값의 차이에 대한 절댓값을 계산하시오.

```
import pandas as pd
california=pd.read_csv("/content/yemoonsaBigdata/datasets/Part3/202_housing.csv")
california
```

```
california80=california[:int(len(california)*0.8)]
print(len(california))
print(len(california80))
```

```
california80['total_bedrooms'].describe()
```

```
med=california80['total_bedrooms'].median()
fill=california80['total_bedrooms'].fillna(med)
std1=california80['total_bedrooms'].std()
std2=fill.std()
print(abs(std1-std2))
```

```
fill.std()
```

2-3.califormia housing 인구컬럼에서 이상값의 합계를 계산하시오.
이상값은 평균에서 1.5*표준편차를 초과하거나 미만인 값의 범위로 정한다.

```
min=california['population'].mean()-california['population'].std()*1.5
max=california['population'].mean()+california['population'].std()*1.5
print(california['population'][(california['population']<min)|(california['population']>max)].sum())
```

3-3.Titanic데이터 세트에서 빈 값 또는 결측치를 확인하여 결측치의 비율이 가장 높은 변수명을 출력하시오

```
titanic=pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/303_titanic.csv')
titanic
```

```
titanic.info()
```

```
isna=titanic.isna().describe()
isna=isna.T
```

```
isna['null']=isna['count']-isna['freq']
isnanull=isna['null']
```

```
isnanull=pd.DataFrame(isnanull)
```

```
print(isnanull.sort_values('null',ascending=False).index[0])
```

```
print(titanic.isnull().sum().index[titanic.isnull().sum().argmax()])
```
