```python
import cv2

video_file = "./img/big_buck.avi" #   

cap = cv2.VideoCapture(video_file) #      ---
if cap.isOpened():                 #    
    while True:
        ret, img = cap.read()      #         --- 
        if ret:                     #   
            cv2.imshow(video_file, img) #    --- 
            cv2.waitKey(25)            # 25ms (40fps )   --- 
        else:                       #     ,
            break                   #  
else:
    print("can't open video.")      #    
cap.release()                       #   
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread('./img/blank_500.jpg')

cv2.line(img, (50, 50), (150, 50), (255,0,0))   #  1 
cv2.line(img, (200, 50), (300, 50), (0,255,0))  #  1 
cv2.line(img, (350, 50), (450, 50), (0,0,255))  #  1 

# (+) 10       
cv2.line(img, (100, 100), (400, 100), (255,255,0), 10)          
# (+) 10       
cv2.line(img, (100, 150), (400, 150), (255,0,255), 10)          
# (+) 10       
cv2.line(img, (100, 200), (400, 200), (0,255,255), 10)          
# (++) 10   
cv2.line(img, (100, 250), (400, 250), (200,200,200), 10)        
#  10     
cv2.line(img, (100, 300), (400, 300), (0,0,0), 10)                    

# 4 
cv2.line(img, (100, 350), (400, 400), (0,0,255), 20, cv2.LINE_4)   
# 8 
cv2.line(img, (100, 400), (400, 450), (0,0,255), 20, cv2.LINE_8)    
#   
cv2.line(img, (100, 450), (400, 500), (0,0,255), 20, cv2.LINE_AA)   
#    
cv2.line(img, (0,0), (500,500), (0,0,255))                      

cv2.imshow('lines', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b, r  Merge
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img2)
plt.show()
```

```python
import cv2
import numpy as np                          #    numpy   ---

img = cv2.imread('./img/blank_500.jpg')

# Numpy array   ---
#    
pts1 = np.array([[50,50], [150,150], [100,140],[200,240]], dtype=np.int32) 
#  
pts2 = np.array([[350,50], [250,200], [450,200]], dtype=np.int32) 
#  
pts3 = np.array([[150,300], [50,450], [250,450]], dtype=np.int32) 
# 5 
pts4 = np.array([[350,250], [450,350], [400,450], [300,450], [250,350]],\
                 dtype=np.int32) 

#   ---
cv2.polylines(img, [pts1], False, (255,0,0))       #    
cv2.polylines(img, [pts2], False, (0,0,0), 10)     # 3    ---
cv2.polylines(img, [pts3], True, (0,0,255), 10)    # 3    ---
cv2.polylines(img, [pts4], True, (0,0,0))          # 5   

cv2.imshow('polyline', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b, r  Merge
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img2)
plt.show()
```

```python
import cv2

file_path = './img/girl.jpg'
img = cv2.imread(file_path)                            #    
img_gray = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE) #    

cv2.namedWindow('origin')                               # origin   
cv2.namedWindow('gray', cv2.WINDOW_NORMAL)              # gray   
cv2.imshow('origin', img)                               # origin   
cv2.imshow('gray', img_gray)                            # gray   

cv2.moveWindow('origin', 0, 0)                          #   
cv2.moveWindow('gray', 100, 100)                        #   

cv2.waitKey(0)                                          #  
cv2.resizeWindow('origin', 200, 200)                    #    ( )
cv2.resizeWindow('gray', 100, 100)                      #    ( ))

cv2.waitKey(0)                                          #  
cv2.destroyWindow("gray")                               # gray  

cv2.waitKey(0)                                          #  
cv2.destroyAllWindows()                                 #   
```

```python
import cv2

img_file = "./img/girl.jpg" 
img = cv2.imread(img_file) 
title = 'IMG'                   #   
x, y = 100, 100                 #  

while True:
    cv2.imshow(title, img)
    cv2.moveWindow(title, x, y)
    key = cv2.waitKey(0) & 0xFF #    , 8 
    print(key, chr(key))        #   ,    
    if key == ord('h'):         # 'h'    
        x -= 10
    elif key == ord('j'):       # 'j'    
        y += 10
    elif key == ord('k'):       # 'k'    
        y -= 10
    elif key == ord('l'):       # 'l'    
        x += 10
    elif key == ord('q') or key == 27: # 'q'  'esc'  
        break
        cv2.destroyAllWindows()
    cv2.moveWindow(title, x, y )   #    
        
```

```python
import cv2

title = 'mouse event'                   #  
img = cv2.imread('./img/blank_500.jpg') #   
cv2.imshow(title, img)                  #   

def onMouse(event, x, y, flags, param): #     ---
    print(event, x, y, )                #  
    if event == cv2.EVENT_LBUTTONDOWN:  #     ---
        cv2.circle(img, (x,y), 30, (0,0,0), -1) #  30      
        cv2.imshow(title, img)          #     ---

cv2.setMouseCallback(title, onMouse)    #    GUI   ---

while True:
    if cv2.waitKey(0) & 0xFF == 27:     # esc 
        break
cv2.destroyAllWindows()
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b, r  Merge
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img2)
plt.show()
```

```python
import cv2

title = 'mouse event'                   #  
img = cv2.imread('./img/blank_500.jpg') #   
cv2.imshow(title, img)                  #   

colors = {'black':(0,0,0),
         'red' : (0,0,255),
         'blue':(255,0,0),
         'green': (0,255,0) } #   

def onMouse(event, x, y, flags, param): #     ---
    print(event, x, y, flags)                #  
    color = colors['black']
    if event == cv2.EVENT_LBUTTONDOWN:  #     ---
        #      
        if flags & cv2.EVENT_FLAG_CTRLKEY and flags & cv2.EVENT_FLAG_SHIFTKEY : 
            color = colors['green']
        elif flags & cv2.EVENT_FLAG_SHIFTKEY : #    
            color = colors['blue']
        elif flags & cv2.EVENT_FLAG_CTRLKEY : #    
            color = colors['red']
        #  30      
        cv2.circle(img, (x,y), 30, color, -1) 
        cv2.imshow(title, img)          #     ---

cv2.setMouseCallback(title, onMouse)    #    GUI   ---

while True:
    if cv2.waitKey(0) & 0xFF == 27:     # esc 
        break
cv2.destroyAllWindows()
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b, r  Merge
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img2)
plt.show()
```

```python
import cv2
import numpy as np

win_name = 'Trackbar'                                   #  

img = cv2.imread('./img/blank_500.jpg')
cv2.imshow(win_name,img)                                #    

#      ---
def onChange(x):                                        
    print(x)                                            #     --- 
    # 'R', 'G', 'B'        --- 
    r = cv2.getTrackbarPos('R',win_name)               
    g = cv2.getTrackbarPos('G',win_name)               
    b = cv2.getTrackbarPos('B',win_name)               
    print(r, g, b)
    img[:] = [b,g,r]                                    #       --- 
    cv2.imshow(win_name, img)                           #    

#      --- 
cv2.createTrackbar('R', win_name, 255, 255, onChange)  
cv2.createTrackbar('G', win_name, 255, 255, onChange)
cv2.createTrackbar('B', win_name, 255, 255, onChange)

while True:
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()        
b, g, r = cv2.split(img)   # img b,g,r 
img2 = cv2.merge([r,g,b]) # b, r  Merge
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img2)
plt.show()
```
